**Spark Reader**  
Reads partner input files into Spark DataFrames based on configuration.

**Spark Writer**  
Writes processed datasets to designated output paths.



```python
def read_partner_file(spark: SparkSession, p: PartnerConfig) -> DataFrame:
    return (
        spark.read.format("csv")
        .option("header", "true" if p.has_header else "false")
        .option("delimiter", p.delimiter)
        .option("mode", "PERMISSIVE")
        .load(p.file_path)
    )
```

```python
def write_csv(df: DataFrame, path: str, coalesce_one: bool, mode: str):
    out = df.coalesce(1) if coalesce_one else df
    out.write.mode(mode).option("header", "true").csv(path)
```
